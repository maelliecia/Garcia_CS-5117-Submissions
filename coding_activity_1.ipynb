Using Python (or some other programming language), plot sampled versions of the sinusoidal signal described as:
$$ x(t) = 9cos(200πt + 0.4π)$$

over the range $-0.02 \leq t \leq 0.02$ seconds. For your output, you should produce three graphs with different sample intervals Ts, namely $T_s = 0.0075$,  $T_s = 0.005$, and  $T_s = 0.0005$.

### Solution:

```python
import numpy as np
import matplotlib.pyplot as plt

# Define sinusoidal signal
def x(t):
    return 9 * np.cos(200 * np.pi * t + 0.4 * np.pi)

# Define time range & sample intervals
t = np.linspace(-0.02, 0.02, 1000)
Ts = [0.0075, 0.005, 0.0005]

# Create subplots
fig, axs = plt.subplots(3, 1, figsize=(10, 12))

# Plot for each sample interval
for i, Ts in enumerate(Ts):
    t_sampled = np.arange(-0.02, 0.02, Ts)
    x_sampled = x(t_sampled)

    axs[i].plot(t, x(t), color='grey', linewidth=0.5)
    axs[i].stem(t_sampled, x_sampled, basefmt='k-')
    axs[i].set_title(f'Samples of Sinusoid: $T_s = {Ts} s$')
    axs[i].set_xlabel('Time $t$ (s)')
    axs[i].set_ylabel('$x(t)$')
    axs[i].grid(True, linestyle=':')

plt.tight_layout()
plt.show()
```
